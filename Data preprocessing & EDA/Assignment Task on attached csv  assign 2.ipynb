## Removing Outliers

```python
## Outliers in hours_per_week
sns.boxplot(df["hours_per_week"])
```

```python
df["hours_per_week"].describe()
```

```python
IQR=77-29
IQR
max_limit=77+1.5*IQR
max_limit
```

```python
df = df[df['hours_per_week'] <= max_limit]

```

```python
sns.boxplot(df["hours_per_week"])
```

```python
sns.boxplot(df["bmi"])
```

```python
df["bmi"].describe()
```

```python
IQR=30.4-23.7
IQR
max_limit=30.4+1.5*IQR
print("max limit ",max_limit)
min_limit=23.7-1.5*IQR
print("min limit ", min_limit)
```

```python
df = df[(df['bmi'] >= min_limit) & (df['bmi'] <= max_limit)]
df
```

```python
sns.boxplot(df["bmi"])
```

```python
df.isnull().sum()
```

```python
## Filling null values
```

```python
df["workclass"] = df["workclass"].fillna(df["workclass"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["occupation"] = df["occupation"].fillna(df["occupation"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["native_country"] = df["native_country"].fillna(df["native_country"].mode()[0])
```

```python
df.dtypes
```

```python
df.head()
```

```python

df["annual_income"] = pd.to_numeric(df["annual_income"], errors="coerce")


median_value = df["annual_income"].median()
df["annual_income"] = df["annual_income"].fillna(median_value)


df["annual_income"] = df["annual_income"].astype(int)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
median_value = df["bmi"].median()
df["bmi"] = df["bmi"].fillna(median_value)
```

```python
df.isnull().sum()
```

```python
df["credit_score"] = pd.to_numeric(df["credit_score"], errors="coerce")


median_score = df["credit_score"].median()


df["credit_score"] = df["credit_score"].fillna(median_score).astype(int)
```

## Encoding categorical variables

```python
df["gender"] = df["gender"].map({"Male": 1, "Female": 2, "Other": 0})
```

```python
df.head()
```

```python
df["education"].unique()
```

```python
df["education"] = df["education"].map({"High School": 1, "Bachelors": 2, "Masters": 3, "PhD":4})
```

```python
df.head()
```

```python
df["marital_status"].unique()
```

```python
df["marital_status"]=df["marital_status"].map({"Single":0, "Married":1, "Divorced":2, "Widowed":3})
```

```python
df.dtypes
```

```python
df.head()
```

```python
df["workclass"].unique()
```

```python
df["workclass"]=df["workclass"].map({"Unemployed":0, "Government":1, "Self-Employed":2, "Private":3})
```

```python
df.head(15)
```

```python
df.dtypes
```

```python
df["occupation"].unique()
```

```python
df["occupation"]=df["occupation"].map({"Other-service":0, "Sales":1, "Tech-support":2, "Prof-specialty":3, "Exec-managerial":4, "Machine-op-inspct":5, "Craft-repair":6})
```

```python
df.head(15)
```

```python
df.dtypes
```

```python
df["native_country"].unique()
```

```python
df["native_country"] = df["native_country"].replace({
    "us": "USA",
    "U.S.": "USA"
})
```

```python
df["native_country"].unique()
```

```python
df["native_country"]=df["native_country"].map({"USA":1, "UK":2, "India":3, "Mexico":4, "Canada":5, "Germany":6})
```

```python
df.head()
```

```python
df["loan_status"].unique()
```

```python
df["loan_status"]=df["loan_status"].map({"denied":0, "approved":1})
```

```python
df["smoking_status"].unique()
```

```python
df["smoking_status"]=df["smoking_status"].map({"occasional":1, "smoker":2, "non-smoker":0})
```

```python
df["physical_activity"].unique()
```

```python
df["physical_activity"]=df["physical_activity"].map({"low":0, "medium":1, "high":2})
```

```python
df.head(30)
```

```python
df = df.drop("id", axis=1)## now i am dropping id column because it is not needed in final dataset 
```
